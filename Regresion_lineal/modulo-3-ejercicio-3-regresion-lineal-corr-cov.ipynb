1. Pregunta teórica: ¿Qué diferencia existe entre la covarianza y la correlación?


>- La `covarianza` indica si ambas variables varían en la misma dirección 
>- **covarianza positiva**: si aumenta el valor de una variable aumenta el variable de la otra 
>- **covarianza negativa**: cuando aumenta el valor de una de las variables, disminuye el de la otra.
>- El valor de la covarianza no es importante, solo el signo (si es positivo o negativo)


>- La `correlación` varía entre -1 y 1. Dos variables están asociadas o correlacionadas cuando una variable nos da información acerca de la otra. La correlación indica la dirección (positiva o negativa) y la "fuerza" de esa relación.
>
>   - Si el valor de la correlación es 0 significa que no hay relación entre las variables. 
>
>    - Si el valor es positivo significa que cuando aumenta el valor de una variable aumenta el de la otra. 
>
>    - Si el valor es negativo significa que cuando disminuye el valor de una variable disminuye el de la otra.


2. Calculad la covarianza y la correlación de nuestro dataset. No nos vale solo con que la calculéis, debéis hacer una interpretación de los resultados obtenidos. Calculad solo la correlación de Pearson.


```python
# Calculamos la covarianza
df.cov()
```

> Por ejemplo:
>
>- En el caso de las cantidades gastadas por producto`mnt...`, la covarianza con `income` es positiva, lo que indica que a mayores ingresos, más consumo, como es de esperar.

>
>- La covarianza ente `income` y `kidhome` es negativa, osea, cuantos más hijos pequeños menos ingresos. Mientras que `income` con `teenhome` es positiva, cuando los hijos crecen aumentan los ingresos del hogar.
>- La covarianza entre `recency` y `mntwines`es positiva, a mayor tiempo transcurrido entre compras, mayor el gasto en todos los productos menos en frutas, `mntfruits`, que muestra una covarianza negativa.

```python
# Calculamos la matriz de correlación para todas nuestras variables, utilizamos el método por defecto, 'pearson'
df.corr(method='pearson')
```

> Con los resultados del heatmap veremos más claras las conclusiones

3. Con los resultados de la correlación del ejercicio anterior, cread un heatmap.


```python
# Creamos una máscara
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))

```

```python
sns.heatmap(df.corr(),
            cmap = "YlGnBu", 
            mask = mask,
           annot = True)
plt.show();
```

> La correlación entre la cantidad total sin vino `mnt_total_sin_vino` y las cantidades gastadas en cada producto es muy alta porque la cantidad total es una suma de todas ellas. Vemos que hay mas correlación con `mntmeatproducts` y deducimos que es donde más dinero se gasta.
> La siguiente correlación con más fuerza es la de la cantidad gastada en carne y ventas por catálogo, interpretamos que es la vía de compra más habitual (`mntmeatproducts` y `numcatalogpurchases`)
> La relación entre las ventas de vino por catálogo y por tienda es la misma (`mntwines` , `numcatalogpurchases` y `numstorepurchases`)
